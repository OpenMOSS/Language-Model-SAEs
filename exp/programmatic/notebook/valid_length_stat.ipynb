```python
import re

# 假设文件名为 log.txt
file_path = "/inspire/hdd/ws-8207e9e2-e733-4eec-a475-cfa1c36480ba/embodied-multimodality/public/zfhe/jiaxing_projects/Language-Model-SAEs/exp/log/merge.txt"

# 初始化两个总和
sum_selected_0 = 0
sum_selected_1 = 0

# 打开并读取文件
with open(file_path, "r") as file:
    for line in file:
        # 检查 selected_source 的值
        if "selected_source: 0" in line:
            # 提取 torch.Size 中的整数列表
            match = re.search(r"torch\.Size\(\[(.*?)\]\)", line)
            if match:
                numbers = list(map(int, match.group(1).split(",")))
                sum_selected_0 += sum(numbers)
        elif "selected_source: 1" in line:
            match = re.search(r"torch\.Size\(\[(.*?)\]\)", line)
            if match:
                numbers = list(map(int, match.group(1).split(",")))
                sum_selected_1 += sum(numbers)

# 输出结果
print(f"selected_source 为 0 时的 torch.Size 中整数的总和是: {sum_selected_0}")
print(f"selected_source 为 1 时的 torch.Size 中整数的总和是: {sum_selected_1}")

print(sum_selected_0 / 3)
print(sum_selected_1 / 3)
print(sum_selected_0 / 3 + sum_selected_1 / 3)
```
